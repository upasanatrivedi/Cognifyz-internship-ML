```python
import pandas as pd
from sklearn.model_selection import train_test_split #to split the data into training and testing
from sklearn.preprocessing import LabelEncoder #to change data into numeric values
from sklearn.metrics import r2_score , mean_squared_error #to evaluate model performance
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('dataset_cognifyz.csv')
df.info()
```

```python
print(df.head())
df.describe()
```

```python
#Checking for missing values
print(df.isnull().sum())
df.shape
```

```python
#dropping columns with null values
df.drop(columns = ['Cuisines'] , inplace = True)
df.shape
```

```python
# Identifying catagorical columns
# Print unique values for each column
for col in df.columns:
    print(f"Unique values for column '{col}': {len(df[col].unique())}")
```

```python
# dropping insignificant columns
required_columns = [col for col in df.columns if col not in ['Restaurant ID','Restaurant Name','Address','Longitude','Latitude','Switch to order menu','Locality','Locality Verbose']]
new_data = df[required_columns]
new_data.shape
```

```python
new_data.info()
```

```python
#binary encoding 2 class data:
binary_cols=['Has Table booking','Has Online delivery','Is delivering now']
for col in binary_cols:
  if col in new_data.columns:
    new_data[col]=new_data[col].map({'Yes':1,'No':0})

new_data.head()
```

```python
#encoding the data
columns_to_label_encode = ['City','Rating color', 'Rating text','Currency']

encoder = LabelEncoder()

for col in columns_to_label_encode:
    if col in new_data.columns:
        new_data[col] = encoder.fit_transform(new_data[col])

print(new_data.head())
```

```python
#separating training and testing data

X = new_data.drop('Aggregate rating', axis=1)
Y = new_data['Aggregate rating']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=45)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of Y_test:", Y_test.shape)
```

```python
# Create a Linear Regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Train the model using the training data
model.fit(X_train, Y_train)

# Make predictions on the testing data
predictions = model.predict(X_test)

# Calculate the Mean Squared Error (MSE)
mse_value = mean_squared_error(Y_test, predictions)

# Calculate the R-squared value (R2)
r2_value = r2_score(Y_test, predictions)

# Print the results
print(f"Mean Squared Error: {mse_value:.2f}")
print(f"Goodness of Fit (R2 Score): {r2_value:.2f}")
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Split the data into training and testing sets to avoid overfitting
train_data, test_data, train_targets, test_targets = train_test_split(
    X, Y, test_size=0.1, random_state=205
)

# Create a Gradient Boosting Regressor model with hyperparameters
model = GradientBoostingRegressor(
    learning_rate=0.01,
    min_samples_split=2,
    min_samples_leaf=0.01
)

# Train the model on the training data
model.fit(train_data, train_targets)

# Make predictions on the testing data
predictions = model.predict(test_data)

# Evaluate the model's performance
mse = mean_squared_error(test_targets, predictions)
r2 = r2_score(test_targets, predictions)

# Print the results
print(f"Mean Squared Error: {mse:.2f}")
print(f"Goodness of Fit (R2 Score): {r2:.2f}")
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define a function to train and evaluate a Decision Tree Regressor model
def train_dt_model(X, Y, test_size=0.15, random_state=110):
    # Split the data into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=random_state)

    # Create a Decision Tree Regressor model with hyperparameters
    dt_model = DecisionTreeRegressor(min_samples_leaf=0.01)

    # Train the model on the training data
    dt_model.fit(X_train, Y_train)

    # Make predictions on the testing data
    predictions = dt_model.predict(X_test)

    # Evaluate the model's performance
    mse = mean_squared_error(Y_test, predictions)
    r2 = r2_score(Y_test, predictions)

    # Print the results
    print(f"Mean Squared Error: {mse:.2f}")
    print(f"Goodness of Fit (R2 Score): {r2:.2f}")
    return dt_model

# Call the function to train and evaluate the model
dt=train_dt_model(X, Y)
```

```python
def get_feature_importances(dt, feature_names):
    importances = dt.feature_importances_
    feature_importances = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
    feature_importances = feature_importances.sort_values(by='Importance', ascending=False)
    return feature_importances

feature_importances = get_feature_importances(dt, dt.feature_names_in_)
print(feature_importances.head(10))
```

The Decision Tree Regressor model achieves the best results on the dataset, making it the most suitable choice for predicting restaurant ratings.
